Structure

````python
model = Sequential()
model.add(Dense(10,activation='relu',input_dim = 4))
model.add(Dense(10,activation='relu'))
model.add(Dense(1,activation='linear'))

model.compile(optimizer = 'adam',loss = 'mse')
model.summary()

````





model = Sequential()


* Sequential() means you are creating a **linear stack of layers** — each layer’s output becomes the next layer’s input.
* It’s the simplest type of neural network model in Keras.

---

###  2️⃣ Add Layers

#### (a) Input + First Hidden Layer

```python
model.add(Dense(10, activation='relu', input_dim=4))
```

* `Dense` → A **fully connected** layer (each neuron connects to all neurons in the previous layer).
* `10` → Number of neurons in this layer.
* `activation='relu'` → Uses the **ReLU (Rectified Linear Unit)** activation function:
  [
  f(x) = \max(0, x)
  ]
  It helps the model learn non-linear patterns.
* `input_dim=4` → The input data has **4 features (columns)** — so the model expects each input sample to have 4 values.

✅ So this is your **input + first hidden layer**.

---

#### (b) Second Hidden Layer

```python
model.add(Dense(10, activation='relu'))
```

* Another fully connected layer with **10 neurons**.
* Again uses **ReLU** to introduce non-linearity.
* Learns more complex relationships between input features.

✅ This is your **second hidden layer**.

---

#### (c) Output Layer

```python
model.add(Dense(1, activation='linear'))
```

* This layer has **1 neuron** → suitable for **regression problems** (predicting a continuous number like car purchase amount 💰).
* `activation='linear'` → No transformation (the neuron just outputs the weighted sum).
  So, it can predict any real value.

✅ This is your **output layer**.

---

### ⚙️ 3️⃣ Compile the Model

```python
model.compile(optimizer='adam', loss='mse')
```

* **`optimizer='adam'`** → Adam optimizer adjusts learning rate automatically; efficient for most problems.
* **`loss='mse'`** → Mean Squared Error; used in regression to measure how far predictions are from real values.

✅ This tells Keras how the model should **learn**.

---

model.summary()

# Model Traning

````python
history = model.fit(x_train,y_train,epochs=10,validation_split=0.2)
````

````python
y_pred = model.predict(x_test)
y_pred = (y_pred>0.5)
print(np.concatenate((y_pred.reshape(len(y_pred),1), y_test.reshape(len(y_test),1)),1   ))
````

y_pred = (y_pred > 0.5)

This converts probabilities into binary values (True/False or 1/0).

The threshold 0.5 means:

If prediction > 0.5 → class = 1 (True)

If prediction ≤ 0.5 → class = 0 (False)

✅ So now y_pred becomes a list of predicted classes instead of probabilities.

````python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Mean Squared Error:", mse)
print("Mean Absolute Error:", mae)
print("R² Score:", r2)
````
